```python
import pandas as pd
```

```python
df = pd.read_csv("replications_database.csv")
```

```python
df['discipline'] = df['discipline'].apply(lambda x: str(x).lower())
```

```python
list(set(df['validated_person']))
```

```python
team_names = ['LK', 'LK, LR', 'CD', 'Veronica Boyce', 'LR', 'LKasseckert']

df['validated_person'] = df['validated_person'].apply(
    lambda x: f"forrt.org team: {x}" if isinstance(x, str) and x in team_names else x
)
```

```python
# Rename columns
df = df.rename(columns={
    'original_doi': 'original_url',
    'replication_doi': 'replication_url'
})

# Add DOI prefix for non-empty string values
for col in ['original_url', 'replication_url']:
    df[col] = df[col].apply(
        lambda x: f"http://doi.org/{x.strip()}" if isinstance(x, str) and x.strip() else x
    )
```

```python
df.to_csv("replications_database.csv")
```
